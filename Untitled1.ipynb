```python
df1 = df[['RUBRICA','LATITUDE','LONGITUDE']] # selecao das colunas desejadas

df1=df1.dropna() # remover dados NaN
df1_check=df.isna()
check_for_any_nan= df1.isna().any().any()
total_nan_values = df1.isna().sum().sum()
print('|Base filtrada, sem NaN nas colunas de coordenadas|\n')
print(f'NaN Presence: {check_for_any_nan}')
print(f'Total Number of NaN values: {total_nan_values:}\n')

info = df1.shape # info tabela
print(f'Linhas | Colunas: {info}\n')

df1.head(5)
```

```python
map = folium.Map(location=[-22.074264, -47.4625256], tiles='cartodb positron', zoom_start=6, width=640, height=480) #map, make_cluster, heatmap, geojson
marker_cluster = MarkerCluster().add_to(map)

for rubrica, lat,lon in zip(df1['RUBRICA'],df1['LATITUDE'],df1['LONGITUDE']): # Posicoes no mapa
        folium.Marker(
        [lat, lon],
        popup= rubrica,
        ).add_to(marker_cluster)

positions = list(zip(df1['LATITUDE'], df1['LONGITUDE'])) # mapa de calor
pos_samp = sample(positions, 247) # max 22k linhas
HeatMap(pos_samp, radius = 15).add_to(marker_cluster)

estado = '/content/drive/My Drive/geojs-35-mun.json' # estado/municipios SP
geo_json_data = json.load(open(estado)) 
style =  {'fillColor': 'blue', 
              'color': 'blue',
             'weight': 0.2, 
               }
folium.GeoJson(geo_json_data, style_function=lambda x:style).add_to(marker_cluster)

map
#mapa.save('map.html')
```

**5: PREPARING THE DATA**

```python
data = df.groupby('DATAOCORRENCIA').size().reset_index() # agrupar informacoes da coluna
dat=data # selecionar tabela data e criar uma nova
dat.columns = ['ds','y'] # renomear colunas
dat.head(5)
```

**5.1: MAKE PREDICTIONS**

```python
m = Prophet() # Forcasting into the future
m.fit(dat)
future_dates=m.make_future_dataframe(periods=365)
forecast = m.predict(future_dates)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#plot_plotly(m, forecast, uncertainty=True)
fig1 = m.plot(forecast, uncertainty=True)
```

```python
#plot_components_plotly(m, forecast, uncertainty=True)
fig2 = m.plot_components(forecast)
```

```python
df_cv = cross_validation(m, initial='730 days', period='180 days',horizon ='365 days')
df_cv.head(5)
```

```python
df_cv[['y', 'yhat']].plot(figsize=(12, 4))
legend(['Dado Real','Previs√£o'])
title('\nCross Validation')
```

**6: DIAGNOSTICS**

```python
df_p = performance_metrics(df_cv)
df_p.head(5)
```

```python
fig3 = plot_cross_validation_metric(df_cv, metric='mape')
```

```python
metric_df = forecast.set_index('ds')[['yhat']].join(dat.set_index('ds').y).reset_index()
metric_df.tail()
metric_df.dropna(inplace=True)
```

```python
r2_score = r2_score(metric_df.y, metric_df.yhat)
mean_squared_error = mean_squared_error(metric_df.y, metric_df.yhat)
mean_absolute_error = mean_absolute_error(metric_df.y, metric_df.yhat)
print(f'|Metric|\nR¬≤(coefficient of determination) regression score function: {r2_score:.2f}\nMean squared error regression loss: {mean_squared_error:.2f}\nMean absolute error regression loss: {mean_absolute_error:.2f}')
```
